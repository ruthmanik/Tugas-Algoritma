# 5.21.7 Pengelompokan data menggunakan Pandas

Data yang sama dapat dikelompokkan untuk analisis lebih lanjut


```python
#pengelompokkan data dengan fungsi mean
#mean() berfungsi menghitung nilai rata-rata dari sebuah kolom numerik pada dataframe
bunga_csv.groupby(["petal width (cm)"]).mean()
```

```python
#pengelompokkan data dengan fungsi mean
#mean() berfungsi menghitung nilai rata-rata dari sebuah kolom numerik di dataframe
#sort_values akan mengurutkan berdasarkan nilai terkecil ke terbesar dari data yang diminta
#ascending=True menentukan pengurutan sort_values dari nilai terkecil ke terbesar
bunga_csv.groupby(["petal width (cm)"]).mean().sort_values("sepal width (cm)" ,ascending=True)
```

```python
#pengelompokkan data dengan fungsi sum
#sum() berfungsi untuk menghitung jumlah dari sebuah kolom numerik pada dataframe
bunga_csv.groupby(["petal width (cm)"]).sum()
```

```python
#pengelompokkan data dengan fungsi count
#count() berfungsi untuk memanpilkan jumlah kolom dan baris pada inisial yang dituju
bunga_csv.groupby(["petal width (cm)"]).count()
```

# 5.22.1 Grafik sederhana dengan Matplotlib
Pada dasarnya Matplotlib berfungsi untuk memplot data menjadi grafik, histogram atau bentuk lainnya untuk mempermudah visualisasi data yang sedang
diolah

```python
#ini adalah inisialisasi variabel berdasarkan kolom pada data yang diminta
x = bunga_csv ["sepal width (cm)"]
y = bunga_csv ["petal width (cm)"]

#plot berfungsi untuk menampilkan data koordinat kedalam garis koordinat 
plt.plot(x,y)

#title berfungsi untuk menamai koordinat
plt.title("Grafik sepal dan petal")

#label berfungsi untuk menamai sumbu yang dimaksud pada koordinat
plt.xlabel("X = panjang sepal")
plt.ylabel("Y = panjang petal")
```

# 5.22.2 Properti Pembuatan Grafik dengan Matplotlib
Untuk memperjelas keterangan pada grafik, ada banyak properti yang bisa dimanfaatkan pada Matplotlib

```python
#ini adalah inisialisasi variabel dengan data variabel sesuai inisialisasi kolom yang dituju
x = bunga_csv ["sepal length (cm)"]
y = bunga_csv ["petal length (cm)"]
x1 = bunga_csv ["sepal width (cm)"]
y1 = bunga_csv["petal width (cm)"]

#pembentukan sumbu koordinat yang berukuran 13 x 13 inch dan resolusi gambar 300 dpi
#figure() berfungsi untuk menentukan gambar dengan format inci
#figsize() berfungsi  untuk menentukan ukuran gambar dengan format inci
#dpi berfungsi untuk menentukan resolusi gambar dalam piksel per inci
plt.figure(figsize=(13,13), dpi=300) 

#label berfungsi untuk memberi label atau judul pada plot
#color berfungsi untuk memberi warna pada plot
#linewidth berfungsi untuk ketebalan garis pada plot
#marker berfungsi untuk menandai titik pada plot
#makersize berfungsi unutuk menentukan ukuran dari marker 
#markeredgecolor berfungsi untuk memberi warna pada tepi marker
#linestyle berfungsi untuk menentukan jenis garis pada plot
plt.plot(x,y, label = "Grafik length" ,  color = "pink" , 
         linewidth = 2 , marker = "*" , markersize = 15 , 
         markeredgecolor = "green" , linestyle="-")

#ini untuk membuat plot dari data pada variabel x1 dan y1
#plot ini tidak memiliki variabel lain 
#seperti label ataupun color sehingga plot ini akan menggunakan nilai default untuk setiap parameter
plt.plot(x1,y1, label = "Garik width", marker = "*" , 
         markersize = 15 , markeredgecolor = "blue" )

#penamaan koordinat dengan fungsi title()
#title() berfungsi untuk memasukan data string dan ukuran font
#fontdict berfungsi untuk menentukan ukuran font
plt.title("Grafik Bunga Iris ", fontdict = {"fontname" : "Sans Serif" , "fontsize" : 12})

#inisialisasi sumbu x dan y dengan penamaan menggunakan data string
#fontdict berfungsi untuk menentukan ukuran font
plt.xlabel("X", fontdict={"fontname" : "Sans Serif" , "fontsize" : 6})
plt.ylabel("Y", fontdict={"fontname" : "Sans Serif" , "fontsize" : 6})

#pembentukan output dari label x dan y
plt.xticks(bunga_csv["sepal length (cm)"])
plt.yticks(bunga_csv["petal length (cm)"])
plt.legend()

#pembentukan output dari label x1 dan y1
plt.xticks(bunga_csv["sepal width (cm)"])
plt.yticks(bunga_csv["petal width (cm)"])
plt.legend()

#penyimpanan figure pada grafik.png dengan resolusi 300 dpi
plt.savefig("grafik.png" , dpi=300)

#show() berfungsi untuk mengeluarkan output figure
plt.show()
```

# 5.22.3 Histogram sederhana dengan Matplotlib

```python
#ini adalah inisialisasi variabel dengan menginput data yang dituju sesuai dengan kolom yang diminta
sepal = bunga_csv["sepal length (cm)"]
petal = bunga_csv["petal length (cm)"]

#namun terdapat operator perbandingan sebagai pengelompokan data kolom yang diminta
sepal1 = bunga_csv["sepal width (cm)"]
petal1 = bunga_csv["petal width (cm)"]

#ini adalah inisialisasi sumbu x dan y 
#fontdict berfungsi untuk menentukan ukuran font
plt.xlabel("Sepal (cm)")
plt.ylabel("Petal (cm)")

#bar() berfungsi untuk membentuk histogram
print(plt.bar(sepal,sepal1))
print(plt.bar(petal,petal1))
```

# 5.22.4 Properti Pembuatan Histogram dengan Matplotlib
Berbagai properti pembuatan histogram dengan menggunakan Matplotlib

```python
#inisialisasi variabel dengan menginput data integer
x = [-5,-3,-1,0,1,3,5,6]
y = [31,13,3,1,1,7,21,31]

#figure() berfungsi untuk menentukan gambar dengan format inci
#figsize() berfungsi  untuk menentukan ukuran gambar dengan format inci
plt.figure(figsize=(3,3))

#inisialisasi variabel dengan fungsi bar() beserta menentukan warnanya
hist = plt.bar(x,y,color='purple')
pola = ["x","x","x","x","/","/","/","/"]

#for berfungsi untuk mengurutkan variabel hist
for bar in hist :
    #set_hatch() berfungsi untuk pola di dalam batang
    bar.set_hatch(pola.pop(0))

#show() berfungsi untuk mengeluarkan output figure
plt.show()
```
